```python
import os
import re
from pydub import AudioSegment
from TTS.api import TTS

# ========= SETTINGS =========
INPUT_TXT = "episode02.txt"
OUTPUT_MP3 = "episode02.mp3"
TMP_DIR = "episode02_chunks"
SPEAKER = "Torcull Diarmuid"
LANGUAGE = "en"
MAX_CHARS = 700
SILENCE_MS = 250
BITRATE = "192k"
# ============================

def split_into_chunks(text, max_chars=700):
    text = text.replace("\r\n", "\n")
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()

    parts = re.split(r"(?<=[\.\!\?])\s+", text.replace("\n", " "))
    chunks = []
    cur = ""
    for p in parts:
        if not p:
            continue
        if len(cur) + len(p) + 1 <= max_chars:
            cur = (cur + " " + p).strip()
        else:
            if cur:
                chunks.append(cur)
            cur = p.strip()
    if cur:
        chunks.append(cur)
    return chunks

if not os.path.exists(INPUT_TXT):
    raise RuntimeError(f"Missing input file: {INPUT_TXT}")

with open(INPUT_TXT, "r", encoding="utf-8") as f:
    text = f.read().strip()
if not text:
    raise RuntimeError("Input text file is empty")

print("Loading XTTS v2...")
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to("cuda")

os.makedirs(TMP_DIR, exist_ok=True)

chunks = split_into_chunks(text, MAX_CHARS)
print(f"Generating {len(chunks)} chunks using speaker: {SPEAKER}")

silence = AudioSegment.silent(duration=SILENCE_MS)
full = AudioSegment.empty()

for i, chunk in enumerate(chunks):
    wav_path = os.path.join(TMP_DIR, f"chunk_{i:04d}.wav")
    tts.tts_to_file(
        text=chunk,
        file_path=wav_path,
        language=LANGUAGE,
        speaker=SPEAKER,
    )
    seg = AudioSegment.from_wav(wav_path)
    full += seg + silence

full.export(OUTPUT_MP3, format="mp3", bitrate=BITRATE)

print("Done:", OUTPUT_MP3)
print("Duration (min):", round(len(full) / 1000.0 / 60.0, 2))
```

```python
import os
import re
import time
from pydub import AudioSegment
from TTS.api import TTS

# ========= CONFIG =========
INPUT_TXT = "episode02.txt"
OUTPUT_MP3 = "episode02.mp3"
TMP_DIR = "episode02_chunks"
SPEAKER = "Viktor Menelaos"
LANGUAGE = "en"
MAX_CHARS = 650
SILENCE_MS = 250
BITRATE = "192k"
# ===========================

def split_into_chunks(text, max_chars=650):
    text = text.replace("\r\n", "\n")
    text = re.sub(r"[ \t]+", " ", text)
    text = re.sub(r"\n{3,}", "\n\n", text).strip()
    parts = re.split(r"(?<=[\.\!\?])\s+", text.replace("\n", " "))
    chunks = []
    cur = ""
    for p in parts:
        if not p:
            continue
        if len(cur) + len(p) + 1 <= max_chars:
            cur = (cur + " " + p).strip()
        else:
            if cur:
                chunks.append(cur)
            cur = p.strip()
    if cur:
        chunks.append(cur)
    return chunks

if not os.path.exists(INPUT_TXT):
    raise RuntimeError(f"Missing input file: {INPUT_TXT}")

with open(INPUT_TXT, "r", encoding="utf-8") as f:
    text = f.read().strip()

print("Loading XTTS v2...")
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to("cuda")

os.makedirs(TMP_DIR, exist_ok=True)

chunks = split_into_chunks(text, MAX_CHARS)
total_chunks = len(chunks)

print(f"Using speaker: {SPEAKER}")
print(f"Total chunks: {total_chunks}\n")

silence = AudioSegment.silent(duration=SILENCE_MS)
full = AudioSegment.empty()

start_total = time.time()

for i, chunk in enumerate(chunks):
    chunk_start = time.time()

    wav_path = os.path.join(TMP_DIR, f"chunk_{i:04d}.wav")

    tts.tts_to_file(
        text=chunk,
        file_path=wav_path,
        language=LANGUAGE,
        speaker=SPEAKER,
    )

    seg = AudioSegment.from_wav(wav_path)
    full += seg + silence

    chunk_time = time.time() - chunk_start
    elapsed = time.time() - start_total
    avg_time = elapsed / (i + 1)
    remaining = avg_time * (total_chunks - i - 1)

    percent = (i + 1) / total_chunks * 100

    print(
        f"[{i+1}/{total_chunks}] "
        f"{percent:5.1f}% | "
        f"Chunk time: {chunk_time:5.2f}s | "
        f"Elapsed: {elapsed/60:5.2f}m | "
        f"ETA: {remaining/60:5.2f}m"
    )

print("\nExporting final MP3...")
full.export(OUTPUT_MP3, format="mp3", bitrate=BITRATE)

total_time = time.time() - start_total

print("\nDone:", OUTPUT_MP3)
print("Total duration (min):", round(len(full) / 1000 / 60, 2))
print("Total generation time (min):", round(total_time / 60, 2))
```

```python
episode02_text = """
Volatility is the central quantity in finance. It governs how assets are priced, how portfolios are allocated, and how risk is managed. And yet, for most of the history of quantitative finance, volatility has been treated as something hidden. A latent variable that must be inferred indirectly from the behavior of returns.

The standard toolkit for this inference was built around GARCH models and stochastic volatility frameworks. These work by fitting parametric structures to daily return data. They capture the well-known tendency of volatility to cluster over time. Calm periods followed by turbulent ones. And they produce forecasts that are often reasonable. But they carry a fundamental limitation. They rely on daily squared returns as their primary signal. A single daily squared return is an extremely noisy measure of the volatility that actually prevailed during that day. The models compensate for this noise by averaging over long histories. They use slowly decaying weighted combinations of past observations. The result is a forecast that adapts sluggishly. When volatility shifts abruptly, the model takes days or even weeks to catch up.

There is a deeper problem as well. Moving beyond a handful of assets becomes very difficult. Multivariate GARCH models suffer from a curse of dimensionality. The number of parameters grows rapidly. Positive definiteness constraints become hard to impose. Estimation becomes fragile. In practice, most serious applications involving large portfolios abandon formal volatility modeling altogether. They fall back on ad-hoc methods like exponential smoothing with a normality assumption. The approach works in the sense that it is feasible. But it leaves a great deal of predictive accuracy on the table.

By the late 1990s, financial markets had generated an enormous and largely untapped resource. Tick-by-tick intraday price data. The question was how to use it. Fitting conventional volatility models directly to intraday returns turned out to be problematic. Intraday data carries its own complications. Strong seasonal patterns tied to market openings and closings. Microstructure noise from bid-ask bounce. Serial dependencies that have nothing to do with the underlying volatility process. Models designed for daily horizons could not easily absorb this information. And models designed for the intraday level struggled to capture the longer-run dynamics. The dynamics that matter most for risk management and portfolio construction.

The insight that changed this situation came from an unexpected direction. Not from building a more complex model. But from recognizing that volatility could be measured rather than modeled. The theoretical foundation was the concept of quadratic variation from continuous-time finance. Under standard no-arbitrage conditions, asset prices are semi-martingales. For any semi-martingale there exists a well-defined quadratic variation process. It captures the cumulative squared variation of the price path. A deep result from stochastic calculus shows that this quadratic variation can be approximated. One sums the squared returns over finer and finer partitions of the trading day. Take intraday returns at a sufficiently high frequency. Add up their cross-products. The result converges to the true integrated volatility over that period.

This is the concept of realized volatility. Instead of treating daily volatility as a hidden state, one constructs it directly. By aggregating high-frequency intraday returns. The daily realized variance is simply the sum of squared intraday returns. Say at thirty-minute intervals over the course of the day. The realized covariance between two assets is the corresponding sum of cross-products.

The practical consequences of this shift are striking. Three empirical regularities emerge when realized volatility is computed from foreign exchange data. First, daily returns divided by their realized standard deviation are approximately Gaussian. The fat tails that dominate raw return distributions largely disappear. They vanish once returns are scaled by a precise measure of the volatility that actually occurred. Second, the distribution of realized volatility itself is right-skewed. But its logarithm is nearly Gaussian. Third, the dynamics of log realized volatility exhibit long memory. Autocorrelations decay very slowly. They follow a hyperbolic pattern rather than the exponential decay of standard autoregressive models.

These three facts together point to a remarkably simple forecasting architecture. Log realized volatilities are approximately Gaussian with long-memory dynamics. So one can model them using a fractionally integrated vector autoregression. A linear Gaussian model applied to observable variables. There is no latent state to filter. There is no complex nonlinear estimation. The model is estimated by ordinary least squares. Equation by equation. After applying a fractional differencing filter to account for the long memory.

The forecasting results are compelling. Across multiple exchange rate pairs and forecast horizons, the realized volatility VAR produces superior forecasts. It outperforms daily GARCH. It outperforms RiskMetrics. It outperforms fractionally integrated GARCH. It even outperforms GARCH models estimated on high-frequency data. The advantage is especially clear in how quickly the forecasts respond to regime shifts. When volatility jumps, the realized volatility measure captures the change immediately. Within the same day. The VAR forecast adjusts on the next step. GARCH forecasts lag behind for days. They are tied to a slowly decaying moving average of past daily squared returns.

The framework extends naturally to density forecasting. Standardized returns are approximately normal. Log realized volatilities are approximately normal. So the unconditional return distribution is well described by a lognormal-normal mixture. Coupling the VAR volatility forecast with this mixture distribution produces well-calibrated density forecasts. The predicted quantiles match realized exceedance frequencies with impressive accuracy. Both in-sample and out-of-sample. Value-at-Risk estimates derived from this approach are reliable across the full range of confidence levels.

Perhaps more importantly, the entire construction scales. Realized covariance matrices are positive definite by construction. This holds whenever the number of assets is smaller than the number of intraday return observations. For thirty-minute sampling, that allows systems of up to forty-eight assets. A dramatic expansion. Multivariate GARCH can handle three or four assets in practice. The machinery of linear time-series modeling becomes applicable to covariance forecasting. It is well understood. Computationally efficient. And easily extended.

The broader contribution is a change in how volatility enters the modeling process. One no longer specifies a parametric model for a latent process. One no longer estimates parameters from noisy daily data. Instead, one first constructs a precise observable measure of volatility from high-frequency data. Then applies standard time-series tools to that measure. The hard work shifts from model specification to measurement. The payoff is forecasts that are both more accurate and simpler to produce. This framework laid the groundwork for much of what followed in high-frequency econometrics. The separation of jump and continuous variation. The development of noise-robust estimators. The practical deployment of large-scale covariance forecasting in portfolio management and risk systems.
"""
```

```python
episode02_text = """
Extreme events in financial markets are rare by definition. That simple fact creates a deep measurement problem. Volatility can be estimated from recent data using something like a GARCH model because there are enough observations day to day to track how dispersion evolves. But tail risk, the probability of truly extreme outcomes, is fundamentally different. Crashes do not happen often enough in any single time series to pin down how their likelihood is changing over time. Trying to estimate a dynamic tail exponent from the history of market returns alone is like trying to track a moving target with a telescope that only opens once a year.

This limitation left researchers in an awkward position. Theoretical models in asset pricing had begun to emphasize the importance of time-varying disaster risk. The rare disaster hypothesis and its extensions showed that fluctuations in the probability of extreme negative shocks could explain puzzles like the equity premium and excess volatility. But testing these models required a reliable measure of how tail risk moves through time. The two existing approaches relied on either options data or high-frequency returns. Both were powerful but constrained. Options data only covered roughly twenty years and could not be applied to cash flow fundamentals. High-frequency methods faced similar horizon limitations.

The breakthrough came from rethinking what data to use. Instead of waiting for enough extreme events to accumulate in the aggregate market return series, the idea was to look across the entire cross-section of individual stocks each month. On any given day, thousands of firms trade. Even if crashes are rare for any single stock, in a large enough cross-section, a meaningful number of firms will experience extreme negative returns in any given month. If the tail risks of individual firms share a common dynamic component, then pooling their crash events reveals the current level of that shared tail risk far more precisely than any single time series ever could.

This is grounded in a specific structural assumption. Individual stock returns in their lower tails follow a power law distribution. The shape of that tail is governed by an exponent that has two parts. One is a firm-specific constant that captures each stock's baseline susceptibility to extreme moves. The other is a time-varying common factor that drives how fat the tails are for everyone at once. Firms can differ in their average tail risk, but the fluctuations in tail fatness move together. The estimation then becomes straightforward. Each month, take all daily returns across all stocks, identify those that fall below the fifth percentile threshold, and apply the Hill estimator to those exceedances. The result is a monthly reading of the common tail risk factor.

Several design choices make this work cleanly. The threshold itself moves with volatility because it is defined as a percentile of the cross-section each period. When markets become more volatile and the entire return distribution widens, the threshold widens too. This means the estimator is not contaminated by changes in volatility that leave the shape of the tail unchanged. The procedure also strips out common return factors before estimation, using Fama-French three-factor residuals, so that the tail measure reflects genuine extreme idiosyncratic events rather than broad market moves.

The resulting tail risk series is remarkably persistent, with a monthly autoregressive coefficient of 0.927. That persistence is not mechanical. Each month's estimate uses entirely fresh data with no overlap. High persistence means that when tail risk rises, it stays elevated. This matters enormously for asset pricing because investors will only adjust their required returns in response to shocks that carry information about the future. A transient blip would not affect equilibrium prices. A persistent shift in the probability of crashes would.

The asset pricing evidence is striking. A one standard deviation increase in tail risk forecasts an annualized increase in excess market returns of about four to four and a half percent, depending on the horizon. This holds at one month, one year, three years, and five years. Out-of-sample, tail risk produces positive forecasting R-squared values at every horizon tested. Most alternative predictors surveyed in the literature, including the dividend-price ratio, book-to-market, and term spread, fail to beat the historical mean out-of-sample. Tail risk is one of the very few that does so reliably.

In the cross-section, stocks with high sensitivity to tail risk earn substantially higher returns than stocks with low sensitivity. Sorting stocks into quintiles based on their estimated tail risk loadings produces a spread in annual three-factor alpha of 5.4 percent. This premium survives controls for momentum, liquidity, idiosyncratic volatility, downside beta, and coskewness. Stocks that effectively hedge tail risk, meaning their prices tend to rise when tail risk increases, earn lower returns. Stocks that suffer when tail risk spikes earn higher returns. The pattern is monotonic and robust.

The options market tells a consistent story. Deep out-of-the-money put options, which are explicit crash insurance contracts, show the strongest sensitivity to the tail risk measure. When tail risk rises, the cost of crash protection increases sharply. The spread in predicted returns between deep out-of-the-money and deep in-the-money short put portfolios widens by over seven percent for a one standard deviation tail risk increase.

Two mechanisms explain why a measure built from individual stock crashes carries information about aggregate prices. The first is mathematical. Power law distributions are stable under aggregation. When individual stock returns have power law tails, the tail of the portfolio return inherits the heaviest individual tail. Common fluctuations in firm-level tail risk therefore directly translate into fluctuations in the probability of market-wide crashes. Empirically, the tail risk measure correlates significantly with option-implied skewness and kurtosis of the S and P 500 index. It also predicts more negatively skewed and more leptokurtic market returns in subsequent months.

The second mechanism operates through the real economy. When tail risk rises across firms simultaneously, the value of waiting increases for every firm facing adjustment costs. Investment and hiring slow as firms exercise their option to delay. This channel connects firm-level tail uncertainty to aggregate economic contractions. Vector autoregression analysis confirms that a tail risk shock precedes significant declines in industrial production, employment, and investment over the following year. These effects are distinct from volatility shocks and persist after controlling for them.

What emerged from this work is a practical and theoretically grounded way to track something that was previously unmeasurable at business-cycle frequencies. It connected the empirical literature on heavy-tailed returns, dating back to Mandelbrot and Fama in the 1960s, with the modern theoretical literature on rare disasters and long-run risks. It showed that the cross-section of firms is not just a source of diversification but a source of information about the state of the economy. And it established that tail risk carries a meaningful price in equity markets, one that is distinct from volatility, liquidity, and other known risk factors. The cross-sectional identification strategy has since been adopted in other contexts, from systemic banking risk to liquidity risk measurement, confirming its value as a general tool for extracting dynamic risk signals from panel data.
"""
```
